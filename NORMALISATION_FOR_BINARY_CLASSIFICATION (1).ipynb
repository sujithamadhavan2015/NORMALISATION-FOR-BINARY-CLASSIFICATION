# Removing less important features

```python
df = df.dropna(subset=['Loan_Status'])
```

```python
df = df.drop(['Loan_ID'],axis=1)
```

```python
num = ['ApplicantIncome','CoapplicantIncome','LoanAmount']
```

```python
df[num] = df[num].fillna(df[num].mean())
```

```python
catg = ['Gender','Area']
df[catg] = df[catg].fillna(df.mode().iloc[0])
```

# Using LabelEncoder from sklearn library to change from categorical column to numerical column.

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df['newGender'] = lb.fit_transform(df['Gender'])
df['newArea'] = lb.fit_transform(df['Area'])
df['newLoan_Status'] = lb.fit_transform(df['Loan_Status'])
```

```python
df.drop('Gender',axis=1,inplace=True)
df.drop('Area',axis=1,inplace=True)
df.drop('Loan_Status',axis=1,inplace=True)
```

```python
df.info()
```

# x and y split

```python
x = df.iloc[:,0:5]
y = df.iloc[:,-1]
```

# we going to perform NORMALISATION to improve the data for the model learning

```python
from sklearn import preprocessing
normalized_x = preprocessing.normalize(x)
```

```python
normalized_x
```

# TRIAN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
normalized_x_train,normalized_x_test,y_train,y_test = train_test_split(normalized_x,y,test_size=0.3,random_state=23)
```

# We use SVM to train the datset

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(normalized_x_train,y_train)
```

```python
ypredict = svm.predict(normalized_x_test)
```

#  Evaluation metrics for the SVM algorithm

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cmsvm = confusion_matrix(y_test,ypredict)
recall = recall_score(y_test,ypredict)
accuracy=accuracy_score(y_test,ypredict)
precision = precision_score(y_test,ypredict)
f1score = f1_score(y_test,ypredict)
print('CONFUSION METRICS','\n',cmsvm)
print()
print('RECALLSCORE    = ',recall)
print('ACCURACYSCORE  = ',accuracy)
print('PRECISIONSCORE = ',precision)
print('    F1SCORE    = ',f1score)
```

# KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(normalized_x_train,y_train)
```

```python
y_predict = knn.predict(normalized_x_test)
```

# Evaluation metrics for the KNN algorithm

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cmknn = confusion_matrix(y_test,y_predict)
recall = recall_score(y_test,y_predict)
accuracy = accuracy_score(y_test,y_predict)
precision = precision_score(y_test,y_predict)
f1score = f1_score(y_test,y_predict)
print('CONFUSIONMETRICS  ','\n',cmknn)
print()
print(' RECALLSCORE   = ',recall)
print('ACCURACYSCORE  = ',accuracy)
print('PRECISIONSCORE = ',precision)
print('   F1SCORE     = ',f1score)
```

# **Logistical regression**

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=0)
log_reg.fit(normalized_x_train,y_train)
```

```python
y_predict = log_reg.predict(normalized_x_test)
```
